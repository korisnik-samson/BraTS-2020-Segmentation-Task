```python
import numpy as np
import matplotlib.pyplot as plt

# Adam optimizers were the best performing optimizers so far
```

```python
def plot_list(statistics: dict):    
    plt.plot(statistics['train_loss'], label='train_loss')
    plt.plot(statistics['valid_loss'], label='valid_loss')
    
    plt.legend(loc='upper right')
    
    plt.xlabel('EPOCH')
    plt.ylabel('LOSS')
    
    plt.title('TRAIN & VALIDATION LOSS')
# plt.grid(False)
```

```python
def plot_metrics(statistics: dict):    
    plt.plot(statistics['fscore'], label = 'fscore')
    plt.legend(loc = "lower right")
    plt.ylabel('score')
    plt.xlabel('epoch')
    plt.title('Fscore')
    
    plt.plot(statistics['iou_score'], label ='iou_score')
    plt.legend(loc = "lower right")
    plt.ylabel('SCORE')
    plt.xlabel('EPOCH')
    plt.title('IOU_SCORE')
    
    plt.plot(statistics['precision'], label ='precision')
    plt.legend(loc = "lower right")
    plt.ylabel('SCORE')
    plt.xlabel('EPOCH')
    plt.title('PRECISION SCORE')
    
    plt.plot(statistics['recall'], label ='recall')
    plt.legend(loc = "lower right")
    plt.ylabel('SCORE')
    plt.xlabel('EPOCH')
    plt.title('RECALL')
```

```python
STATS = np.load(f'model/adam/model.npy', allow_pickle=True).item()
plot_list(STATS)
```

```python
plot_metrics(STATS)
```

```python
def get_accuracy(stat_array: any, metric: str):
    accuracy = np.array(stat_array[metric])
    return f'{metric}: {round(accuracy.max(), 3) * 100}%'
```

```python
print(f'IoU Score: {get_accuracy(STATS, "iou_score")}')
print(f'F_Score Score: {get_accuracy(STATS, "fscore")}')
```
